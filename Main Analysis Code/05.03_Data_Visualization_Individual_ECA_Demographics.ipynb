```python
# Create a Dash application
app3 = dash.Dash(__name__)

# Define school levels and x-axis options
school_levels = ['Elementary School', 'Middle School', 'High School']
x_axis_options = ['per_white', 'per_asian', 'per_africanamerican', 'per_hispanic', 'per_other',
                  'per_whiteasian', 'per_blackorhisp'
]

# Define corresponding labels for the x-axis options
x_axis_labels = [
    'White', 'Asian', 'Black', 'Hispanic', 'Other',
    'White & Asian', 'Black & Hispanic'
]

# Combine options with labels for the dropdown
x_axis_dropdown_options = [{'label': label, 'value': value} for label, value in zip(x_axis_labels, x_axis_options)]

# Define a dictionary to map options to updated labels
updated_labels = {
    'per_whiteasian': 'White and Asian',
    'per_blackorhisp': 'Black and Hispanic'
}

# Define the layout of the dashboard
app3.layout = html.Div([
    html.H1("Individual Extracurricular Activities Available"),
    dcc.Dropdown(
        id='outcome-dropdown',
        options=[
            {'label': 'Baseball', 'value': 'EC_SPORTS_BASEBALL'},
            {'label': 'Basketball', 'value': 'EC_SPORTS_BSKTBALL'},
            {'label': 'Bowling', 'value': 'EC_SPORTS_BOWLING'},
            {'label': 'Competitive Cheer', 'value': 'EC_SPORTS_COMPCHEER'},
            {'label': 'Cross Country', 'value': 'EC_SPORTS_CROSSCNTRY'},
            {'label': 'Diving', 'value': 'EC_SPORTS_DIVING'},
            {'label': 'Football', 'value': 'EC_SPORTS_FOOTBALL'},
            {'label': 'Golf', 'value': 'EC_SPORTS_GOLF'},
            {'label': 'Gymnastics', 'value': 'EC_SPORTS_GYMNAST'},
            {'label': 'Ice Hockey', 'value': 'EC_SPORTS_ICEHOCKEY'},
            {'label': 'Other Sports', 'value': 'EC_SPORTS_OTHSPRTS'},
            {'label': 'Powerlifting', 'value': 'EC_SPORTS_POWERLIFT'},
            {'label': 'Skiing', 'value': 'EC_SPORTS_SKIING'},
            {'label': 'Softball', 'value': 'EC_SPORTS_SOFTBALL'},
            {'label': 'Soccer', 'value': 'EC_SPORTS_SOCCER'},
            {'label': 'Swimming', 'value': 'EC_SPORTS_SWIM'},
            {'label': 'Tennis', 'value': 'EC_SPORTS_TENNIS'},
            {'label': 'Track and Field', 'value': 'EC_SPORTS_TRCKFIELD'},
            {'label': 'Volleyball', 'value': 'EC_SPORTS_VOLBALL'},
            {'label': 'Wrestling', 'value': 'EC_SPORTS_WRSTL'},
            {'label': 'Honor Society', 'value': 'EC_CLUBS_HONSOC'},
            {'label': 'Literature and Humanities', 'value': 'EC_CLUBS_LITHUM'},
            {'label': 'Robotics', 'value': 'EC_CLUBS_ROBOTICS'},
            {'label': 'STEM', 'value': 'EC_CLUBS_STEM'},
            {'label': 'Music', 'value': 'EC_CLUBS_MUSIC'},
            {'label': 'Visual Arts', 'value': 'EC_CLUBS_VISARTS'},
            {'label': 'National or Statewide Leadership Association', 'value': 'EC_CLUBS_NSLEAD'},
            {'label': 'Peer Tutoring and Mentoring', 'value': 'EC_CLUBS_PEERTUTMEN'},
            {'label': 'Politics and Government', 'value': 'EC_CLUBS_POLGOV'},
            {'label': 'School Community Organizations', 'value': 'EC_CLUBS_SCHOOLCOM'},
            {'label': 'Speech and Debate', 'value': 'EC_CLUBS_SPCHDEB'},
            {'label': 'Vocational', 'value': 'EC_CLUBS_CARFOC'},
        ],
        value='EC_SPORTS_BASEBALL'  # Default value
    ),
    dcc.Dropdown(
        id='school-level-dropdown',
        options=[{'label': level, 'value': level} for level in school_levels],
        value='Elementary School'  # Default value
    ),
    dcc.Dropdown(
        id='x-axis-dropdown',
        options=x_axis_dropdown_options,
        value='per_white'  # Default value
    ),
    dcc.Graph(id='scatter-plot')
])

@app3.callback(
    dash.dependencies.Output('scatter-plot', 'figure'),
    [dash.dependencies.Input('school-level-dropdown', 'value'),
     dash.dependencies.Input('x-axis-dropdown', 'value'),
     dash.dependencies.Input('outcome-dropdown', 'value')]
)
def update_scatter_plot(selected_school_level, selected_x_axis, selected_size):
    # Filter data based on selected school level
    selected_data = MIGENEDCOMPLETEWITHECA[MIGENEDCOMPLETEWITHECA['School Level'] == selected_school_level]
    
    # Extract selected columns
    y = selected_data['per_ecodis']
    x = selected_data[selected_x_axis]
    color = selected_data[selected_size]

    # Update x-axis label if it's one of the modified options
    x_axis_label = selected_x_axis
    if selected_x_axis in updated_labels:
        x_axis_label = updated_labels[selected_x_axis]

    # Define the title based on the selected outcome
    outcome_title = {
        'EC_SPORTS_BASEBALL': 'Baseball',
        'EC_SPORTS_BSKTBALL': 'Sports',
        'EC_SPORTS_BOWLING': 'Bowling',
        'EC_SPORTS_COMPCHEER': 'Competitive Cheer',
        'EC_SPORTS_CROSSCNTRY': 'Cross Country',
        'EC_SPORTS_DIVING': 'Diving',
        'EC_SPORTS_FOOTBALL': 'Football',
        'EC_SPORTS_GOLF': 'Golf',
        'EC_SPORTS_GYMNAST': 'Gymnastics',
        'EC_SPORTS_ICEHOCKEY': 'Ice Hockey',
        'EC_SPORTS_OTHSPRTS': 'Other Sports',
        'EC_SPORTS_POWERLIFT': 'Powerlifting',
        'EC_SPORTS_SKIING': 'Skiing',
        'EC_SPORTS_SOFTBALL': 'Softball',
        'EC_SPORTS_SOCCER': 'Soccer',
        'EC_SPORTS_SWIM': 'Swimming',
        'EC_SPORTS_TENNIS': 'Tennis',
        'EC_SPORTS_TRCKFIELD': 'Track and Field',
        'EC_SPORTS_VOLBALL': 'Volleyball',
        'EC_SPORTS_WRSTL': 'Wrestling',  
        'EC_CLUBS_HONSOC': 'Honor Society',  
        'EC_CLUBS_LITHUM': 'Literature and Humanities',  
        'EC_CLUBS_ROBOTICS': 'Robotics',  
        'EC_CLUBS_STEM': 'STEM',      
        'EC_CLUBS_MUSIC': 'Music',      
        'EC_CLUBS_VISARTS': 'Visual Arts',
        'EC_CLUBS_NSLEAD': 'National or Statewide Leadership Association',
        'EC_CLUBS_PEERTUTMEN': 'Peer Tutoring and Mentoring',
        'EC_CLUBS_POLGOV': 'Politics and Government',
        'EC_CLUBS_SCHOOLCOM': 'School Community Organizations',
        'EC_CLUBS_SPCHDEB': 'Speech and Debate',
        'EC_CLUBS_CARFOC': 'Vocational',
    }
    
    # Map '0' to 'No' and '1' to 'Yes'
    color_labels = {0: 'Unavailable', 1: 'Available'}
    color = color.map(color_labels)
    
    # Create scatter plot with discrete colors for 0 and 1
    fig3 = px.scatter(
        x=x, y=y, color=color,
        color_discrete_map={'Unavailable': 'red', 'Available': 'blue'},
        labels={'x': f'% {x_axis_label.replace("per_", "").title()}',
                'y': '% Economically Disadvantaged'},
        title=f'{outcome_title[selected_size]}: {selected_school_level}' 
    )
    # Update legend title and colors
    fig3.update_traces(showlegend=True)
    fig3.update_layout(legend_title_text='Legend')
    
    # Update x-axis category order and labels
    return fig3

# Run the application
if __name__ == '__main__':
    app3.run_server(debug=True)
```
